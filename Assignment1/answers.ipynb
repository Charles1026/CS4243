```python
import numpy as np
from PIL import Image 
import cv2
import helpers


img = np.asarray(Image.open("corrupted.png"))

# Plot Original Image Histogram
helpers.plotAndSaveHistogram(img, "Original Image Histogram", path = "./1./")

# As the equalised image is still noisy, we apply a median blur to get rid of the noise
blurredImg = cv2.medianBlur(img, ksize = 9)
helpers.plotAndSaveHistogram(blurredImg, "Blurred Image Histogram", path = "./1./")
Image.fromarray(blurredImg).save("./1./blurred.png")

# As histogram is skewed to both ends, we equalise it to give the image better contrast
equalisedImg = cv2.equalizeHist(blurredImg)
helpers.plotAndSaveHistogram(equalisedImg, "Equalised Image Histogram", path = "./1./")

helpers.show_imgs([img, blurredImg, equalisedImg], ["Raw", "Blurred", "Equalised"])

Image.fromarray(equalisedImg).save("./1./uncorrupted.png")
```
